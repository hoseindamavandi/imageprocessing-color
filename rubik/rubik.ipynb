```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("rubix.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        if image[i, j][0] < 50 and image[i, j][1] > 150 and image[i, j][2] > 150:
            image[i, j] = [255, 0, 0]
            
        elif image[i, j][0] > 150 and image[i, j][1] < 50 and image[i, j][2] > 150:
            image[i, j] = [0, 255, 0]
        
        elif image[i, j][0] > 150 and image[i, j][1] > 150 and image[i, j][2] < 50:
            image[i, j] = [0, 0, 255]
            

```

```python
plt.imshow(image)
```
